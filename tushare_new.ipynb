7、美股数据

```python
import time
import sys
from datetime import datetime, timedelta,date
```

```python
def n_days_ago(n):
	#now=datetime.now()
	today=date.today()
	#print("Now:\n"+str(today))
	ndays_ago=today-timedelta(n)
	#print(str(n)+" days ago:\n"+str(ndays_ago))
	return str(ndays_ago)
```

```python
n=180
while n>0:
    ndate=n_days_ago(n)
    df = ts.tick('300676', conn=cons, date=ndate)
    if df is None:
        print(ndate + ": NULL")
    else:
        print(ndate + ": Valid")
        print(len(df.datetime))
    n=n-1
```
